```python
import numpy as np
import matplotlib.pyplot as plt
from E515 import E515  # Assuming E515 class is saved in E515.py and imported here
```

```python
# Initialize the E515 piezo stage
piezo_stage = E515()

# Function to simulate SHG experiment
def simulate_shg(piezo_stage, start, stop, step):
    positions = np.arange(start, stop, step)
    intensities = np.random.rand(len(positions))  # Random SHG intensities

    # Move the piezo stage and collect data
    shg_data = []
    for pos in positions:
        piezo_stage.set_pos(pos, channel=1)  # Move piezo stage to new position
        intensity = intensities[pos]
        shg_data.append((pos, intensity))
    
    return shg_data

# Parameters for the SHG experiment
start_pos = 0
stop_pos = 100
step_size = 1

# Simulate the SHG experiment
shg_results = simulate_shg(piezo_stage, start_pos, stop_pos, step_size)

# Plotting the SHG intensity versus position
positions, intensities = zip(*shg_results)
plt.plot(positions, intensities, marker='o', linestyle='-', color='b')
plt.title('Second Harmonic Generation Intensity vs Position')
plt.xlabel('Position (um)')
plt.ylabel('Intensity (a.u.)')
plt.grid(True)
plt.show()

# Close the connection to the piezo stage
piezo_stage.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate SHG experiment
def simulate_shg(start, stop, step):
    positions = np.arange(start, stop, step)
    intensities = np.random.rand(len(positions))  # Random SHG intensities

    # Simulate data collection
    shg_data = list(zip(positions, intensities))
    
    return shg_data

# Parameters for the SHG experiment
start_pos = 0
stop_pos = 100
step_size = 1

# Simulate the SHG experiment
shg_results = simulate_shg(start_pos, stop_pos, step_size)

# Plotting the SHG intensity versus position
positions, intensities = zip(*shg_results)
plt.plot(positions, intensities, marker='o', linestyle='-', color='b')
plt.title('Simulated Second Harmonic Generation Intensity vs Position')
plt.xlabel('Position (arbitrary units)')
plt.ylabel('Intensity (arbitrary units)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the simulation
num_data_points = 100  # Number of data points to simulate
energy_min = 1.9       # Minimum photon energy in eV
energy_max = 2.5       # Maximum photon energy in eV

# Generate random photon energies within the specified range
photon_energies = np.random.uniform(energy_min, energy_max, num_data_points)

# Generate random SHG intensities
# For simplicity, we assume the intensity is proportional to the square of the energy
shg_intensities = np.random.rand(num_data_points) * photon_energies**2

# Normalize the intensities for better visualization
shg_intensities_normalized = shg_intensities / np.max(shg_intensities)

# Plot the simulated SHG intensity versus photon energy
plt.scatter(photon_energies, shg_intensities_normalized, color='r')
plt.title('Simulated SHG Intensity vs Photon Energy')
plt.xlabel('Photon Energy (eV)')
plt.ylabel('Normalized Intensity (a.u.)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the simulation
num_samples = 1000  # Number of samples to simulate
wavelength_min = 400  # Minimum input wavelength in nm
wavelength_max = 700  # Maximum input wavelength in nm

# Generate random input wavelengths within the specified range
input_wavelengths = np.random.uniform(wavelength_min, wavelength_max, num_samples)

# Calculate the output wavelength for SHG (half the input wavelength)
output_wavelengths = input_wavelengths / 2

# Generate random SHG intensities
# For simplicity, we assume the intensity is proportional to the square of the electric field
shg_intensities = np.random.rand(num_samples) * (1 / output_wavelengths)**2

# Normalize the intensities for better visualization
shg_intensities_normalized = shg_intensities / np.max(shg_intensities)

# Plot the simulated SHG intensity versus output wavelength
plt.scatter(output_wavelengths, shg_intensities_normalized, color='g', alpha=0.5)
plt.title('Simulated SHG Intensity vs Output Wavelength')
plt.xlabel('Output Wavelength (nm)')
plt.ylabel('Normalized Intensity (a.u.)')
plt.grid(True)
plt.show()
```
